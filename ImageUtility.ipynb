Image classification

```python
from cv2 import *
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline 
```

```python
def load_image(path):
    img = cv2.imread(path)
    plt.figure()
    plt.imshow(img)
    plt.figure()
    img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    plt.imshow(img)
    return img
```

```python
img = load_image("data/stripe.jpg")
```

1- Vertical Edge Detection

```python
from scipy import signal

def vertical_edge(img):
    filt = np.array([[1,0,-1],[1,0,-1],[1,0,-1]])
    img = signal.convolve2d(img,filt,"same")
    plt.figure()
    plt.imshow(img)
    return img

def sobel_filter(img):
    filt = np.array([[1,0,-1],[2,0,-2],[1,0,-1]])
    img = signal.convolve2d(img,filt,"same")
    plt.figure()
    plt.imshow(img)
    return img

def scharr_filter(img):
    filt = np.array([[3,0,-3],[10,0,-10],[3,0,-3]])
    img = signal.convolve2d(img,filt,"same")
    plt.figure()
    plt.imshow(img)
    return img
```

```python
img_after_convolve_1 = vertical_edge(img)
img_after_convolve_2 = sobel_filter(img)
img_after_convolve_3 = scharr_filter(img)
```
